# オープンデータの利用と可視化
## 準備

````python
#マジックファンクション
%matplotlib inline
#お馴染みpandasを読み込み
import pandas as pd
````

````python
#このアドレスは期間限定なので注意（横浜市が情報を更新したら変更する必要があります。）
#ローカルに保存した方が安心ですね
open_data = "http://www.city.yokohama.lg.jp/ex/stat/opendata/suikei01/e1yokohama1803.csv"
print(open_data)
````

````python
#市区名をインデックスにしてpandas形式のオブジェクトを作成
df = pd.read_csv(open_data, index_col='市区名')
#今からこれを解析します
df
````

# オープンデータの利用と可視化
## 準備

````python
#リネーム（項目名の変更）も可能
df.rename(columns={'全国地方公共団体コード':'Code'}, inplace=True)
df
````

````python
#末尾だけ見ることもできる!
df.tail(3)
````

# オープンデータの利用と可視化
## 人口データをpandasで抽出

````python
#上の表だと使わないデータが多いので、必要なものを抜き出します。
key_data = ['人口総数［人］','男［人］','女［人］']
# 横浜市(コード141003)を覗く区のデータのなかで、人口総数、男性の人数、女性の人数を含むカラムだけを取得
pp = df[df.Code !=141003][key_data]
pp
````

# オープンデータの利用と可視化
## 人口データをグラフ化

````python
pp.plot(kind='bar', figsize=(12,6), y=["男［人］","女［人］"], fontsize=17)
````

# オープンデータの利用と可視化
## データを並べ換える

````python
#人口総数[人]のカラムに含まれる値でソートする。ascendingをTrueにすると昇順になる。
sorted_pp = pp.sort_values(by="人口総数［人］", ascending=False)
sorted_pp
````

# オープンデータの利用と可視化
## 並べ替えたデータを表示する

````python
#並べ替えたデータを表示。どの区が人口が多いとかがわかりやすい！
sorted_pp.plot(kind='bar', figsize=(12,6), y=['男［人］', '女［人］'], fontsize=17)
````
