## 問題:y=sin(x) ( -2π < x < 2π )のグラフを作成してください。

# 答え

```python
# まずデータを作成します
import scipy.constants
import numpy as np

pi = scipy.constants.pi
step = 0.001

x = np.arange(-2*pi, 2*pi, step)
print(x)
```

```python
y = [np.sin(x_fragment) for x_fragment in x]
print(y)
```

# 答え

```python
# 　前のステップで作成したデータを使ってグラフを作成します。
from matplotlib import pyplot
pyplot.figure(figsize=(15, 10))
pyplot.plot(x,y)
```
